1 . Inheritance ! 



```python
## creating a Pparent class 
class shape :
    def  __init__(self,color):
        self.color = color 
    def area (self):
        return None 
```

```python
## creating a child class !

class Rectangle(shape):
    def __init__(self , color , width , height ):
        super().__init__(color)
        self.width = width 
        self.height = height 
    def area (self):
        return self.width * self.height
```

```python
##creating another class !
class circle(shape):
    def __init__(self,color,radius):
        super().__init__(color)
        self.radius = radius
    def area (self):
        return 3.14 * self.radius **2 
```

```python
## Testing the class ! 
rect = Rectangle("Red",4,5)
print(rect.area())
```

```python
## testing number 2 

circlearea = circle("Blue",5)
print(circlearea.area())
```

2 . Encapulation and data hiding !


Encapsulation is the bundling of data
(attributes) and methods that operate on that data within a 
single unit (class). Data hiding is the mechanism of protecting
certain data from being accessed and modified from outside the class.

```python
## implementing a encapsulation !

class BankAccount:
    def __init__(self, account_number, balance):
        self._account_number = account_number 
        self.__balance = balance              
#methods 
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print("Deposit successful")
        else:
            print("Deposit invalid")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            print("Withdrawal successful")
        else:
            print("Insufficient balance")

    def get_balance(self):
        return self.__balance

    def _display_account_number(self):
        return self._account_number 

```

```python
account = BankAccount("123987213",0.0)

# deposit 
account.deposit(1000)
account.withdraw(500)

print("Balance:", account.get_balance())  

print("Account Number:", account._display_account_number())  
```

3 .Polymorphism

Polymorphism means "many forms." In object-oriented programming, it refers to the ability of different classes to respond to the same method call in their own way. This allows you to write more flexible and generic code

```python
class Animal : 
    def __init__(self,name):
        self.name = name 

    def speak (self):
        return "the animal sound"
class Dog(Animal):
     def speak(self):
         return "woof"
class cat(Animal):
    def speak(self):
        return "meow"
```

```python
animal = Animal("generic animal")
dog = Dog("Buddy")
cat = cat("Whiskers")
print(animal.speak())
print(dog.speak())
print(cat.speak())
```

4 . Abstraction !!!

Abstraction is the process of hiding complex implementation details and showing only the necessary information to the user. It simplifies the object by exposing only what is relevant.

```python
# implementing Abstraction ! 

from abc import ABC, abstractmethod

import math 

class shape(ABC):
    @abstractmethod
    def area(self):
        raise NotImplementedError("Subclasses must implemented the area method")
class Rectangle(shape):
    def __init__(self,width,height):
        self.width = width 
        self.height = height 
    def area(self):
        return self.width * self.height 

class circle (shape):
    def __init__(self,radius):
        self.radius = radius 
    def area (self):
        return math.pi * self.radius**2 

```

```python
## we will create instance or you can say it object ! 

rect = Rectangle(4,5)
print(rect.area())

circlearea = circle (3)
print(circlearea.area())
```
