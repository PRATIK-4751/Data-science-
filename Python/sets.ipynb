```python
## find the union and intersection of two sets  !
def union_inter(set1,set2):
    union_ = set1.union(set2)
    inter = set1.intersection(set2)
    return union_,inter
set1 = {1,2,3,4,5}
set2 = {4,5,6,7,8}
print(union_inter(set1,set2))
```

```python
## check if two strings are anagrams of each other ! using sets !
def anagrams(str1,str2):
    set1 = set(str1)
    set2 = set(str2)
    if set1 == set2:
        return True
    else:
        return False
str1 = 'listen'
str2 = 'silent'
print(anagrams(str1,str2))
```

```python
#Find the longest substring of a string with all unique characters using set!
def longest(s):
    char_set = set()
    left = 0 
    max_len = 0 
    for right in range(len(s)):
        while s[right] in char_set:
            char_set.remove(s[left])
            left+=1 
        char_set.add(s[right])
        max_len = max(max_len, right-left+1)
    return max_len

s = 'abccba'
longest(s)
```

```python
#Find all unique pairs of numbers in a list whose sum equals a target using set !
def find_pairs(nums,target):
    pairs = set()
    seen  = set()
    for num in nums:
        complement = target - num 
        if complement in seen : 
            pairs.add((min(num,complement),max(num,complement)))
        seen.add(num)
    return pairs


nums = [1,2,3,4,5,6,7,8,9]
target = 10
find_pairs(nums,target)
```

```python
#Given a list of numbers, find the first duplicate using a set.
def first_duplicate(nums):
    seen=set()
    for num in nums : 
        if num in seen :
            return num 
        seen.add(num)
    return None
#eg
nums = [1,2,3,4,5,6,7,8,9,1]
first_duplicate(nums)
```
