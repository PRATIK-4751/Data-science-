```python
## count the frequency of each element in a list and store it in a dictonary !
def frequency(lst):
    freq = {}
    for i in lst: 
        if i in freq : 
            freq[i] += 1 
        else : 
            freq[i]  = 1
    return freq 
lst = [1,2,3,4,5,6,7,8,9,1,2,3,4,5,6,7,8,9]
print(frequency(lst))
```

```python
## create a dictonary from two list : one keys and one of values 
def create_dict(keys,values):
    dict = {}
    for i in range(len(keys)):
        dict[keys[i]] = values[i]
    return dict 

keys = ['a','b','c','d']
values = [1,2,3,4]
print(create_dict(keys,values))
```

```python
## create if two dictonaries are equal (same keys and values )
def equal_dict(dict1 , dict2 ): 
    if len(dict1)!= len(dict2):
        return False
    for key , value in dict1.items():
        if key not in dict2 and value!= dict2[key]:
            return False 
         
    return True 

dict1= {'a':1,'b':2,'c':3}
dict2 = {'a':1,'b':2,'c':3}
print(equal_dict(dict1,dict2))
```

```python
## merge two dictonaries  if a key exist in both , sum their values 
def merge(dict1, dict2):
    merged_dict = dict1.copy()
    for key , value in dict2.items():
        if key in merged_dict:
            merged_dict[key] += value
        else:
            merged_dict[key] = value

    return merged_dict
dict1 = {'a':1,'b':2,'c':3}
dict2 = {'b':3,'c':4,'d':5}
print(merge(dict1,dict2))
```

```python
## find the key with maxm value in a dictonary ! 
def maxm(dict1):
    maxm = float('-inf')  
    max_key = None
    for key, value in dict1.items():
        if value > maxm:
            maxm = value
            max_key = key
    return max_key  

dict1 = {"a": 1, "b": 2, "c": 3}
print(maxm(dict1))
```

```python
## given a dictomnary of student name and scores , return a list of students who scored above the average 

def above_average(students):
    total = 0 
    for score in students.values():
        total += score 
    avg = total / len(students)
    above = []
    for name,score in students.items():
        if score > avg : 
            above.append(name)
    return above
#eg 

students = {'Praik':80,'Alex':90,'Lucifero':70,'Satan':85}
above_average(students)
```

```python
### Remove all keys from a dictionary where the value is None or an empty string ! 
def remove_empty(d):
    for key in list(d.keys()):
        if d[key] is None or d[key] == '':
            del d[key]
    return d 

#eg 
d = {'a':1,'b':None,'c':''}
remove_empty(d)
```

```python
## groups that are anagrams to a dictionary 
def anagrams(lst):
    d = {}
    for word in lst : 
        sorted_word = ''.join(sorted(word))
        if sorted_word in d : 
            d[sorted_word].append(word)
        else:
            d[sorted_word] = [word]
    return list(d.values())

#eg 
lst = ['act','cat','tac','god','dog','odg']
anagrams(lst)
```

```python
## flatten a nested dictionary (covert multi -level dicts into a single level with a compound keys  !)

def flatten_dict(d,parent_key='',sep='_'):
    items = []
    for k,v in d.items():
        new_key = parent_key + sep + k if parent_key else k
        if isinstance(v,dict):
            items.extend(flatten_dict(v,new_key,sep).items())
        else:
            items.append((new_key,v))
    return dict(items)

nested_dict = {'a':1,'b':{'c':2,'d':{'e':3}}}
print(flatten_dict(nested_dict))
```
