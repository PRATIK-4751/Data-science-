```python
# find the max and min in list without using max and min built in fn 
def max_min(lst):
    max = (lst[0])
    min = (lst[0])
    for i in lst : 
        if i > max: 
            max = i 
        if i < min : 
            min = i 
    return max , min 

lst = [1,2,3,43,4,4,4,5,6,6,6,76,77]
max_min(lst)
```

```python
## remove all duplicates from list while keeping orignal order 
def remove_duplicate(lst): 
    new_lst = []
    for i in lst : 
        if i not in new_lst: 
            new_lst.append(i)
    return new_lst 


##eg 

lst = [12,3,3,4,4,4,45,55,5,5,6,6,6,7,7,7]
remove_duplicate(lst)
```

```python
## rotate list in right by k steps 
def rotated_lst(lst,k):
    k = k  % len(lst)
    res= lst[-k:]+lst[:-k]
    return res
    ##eg 
lst = [1,2,3,4,5]
k = 2
rotated_lst(lst,k)
```

```python
## merge 2 sorted list without using sorted function 
def merge(lst1,lst2):
    lst3 = lst1+lst2 
    for i in range(len(lst3)):
        for j in range(len(lst3)-i-1):
            if lst3[j] > lst3[j+1]:
                lst3[j],lst3[j+1] = lst3[j+1],lst3[j]
    return lst3

lst1 = [1,2,5,7,6]
lst2 = [3,4,8,9,10]
merge(lst1,lst2)
```

```python
## two sum in hash map  !
def two_sum(nums,target):
 hash_map = {}
 for i in range(len(nums)):
    diff = target -nums[i]
    if diff in hash_map:
        return [hash_map[diff],i]
    else:
        hash_map[nums[i]] = i
##eg 

nums = [2,7,11,15]
target = 9
two_sum(nums,target)
```

```python
## implement a function to flatten a nested list 
def flatten(nested_lst):
    flattend_list = []

    for i in nested_lst:
        if isinstance(i,list):
            flattend_list.extend(flatten(i))
        else:
            flattend_list.append(i)
    return flattend_list
print(flatten([1,2,3,[4,5,6],[7,8,9]]))
```

```python
## write a program to find the subarray with maxm sum (kadane's algorithm)

## in this algo we are using the concept of dynamic programming to first find the current_max using the sum of current_index  + current index and then max_sum_so far 
## if the current_max is greater than max_sum_so_far then we update the max_sum_so_far with the current_max
## if the current_max is less than 0 then we update the current_max with 0
## we return the max_sum_so_far at the end of the loop

def kadane(arr):
    max_so_far = arr[0]
    current_max = arr[0]
    for i in arr[1:]: 
        current_max = max(current_max + i, i)
        max_so_far = max(max_so_far, current_max)
    return max_so_far

arr = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
print(kadane(arr))  

```

```python
## implement a function that removes the nth largest element from a list 
def remove(lst,n): 
    sorted_lst = sorted(lst, reverse=True)
    nth_lagest = sorted_lst[n-1]
    lst.remove(nth_lagest)
    return lst, nth_lagest
lst = [1,2,3,4,5,6,7,8,9]
print(remove(lst,5))
```
