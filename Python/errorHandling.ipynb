```python
## open a file and doesn't exist -catch the filenotfounderror and print a custom message!
try:
    file = open("hello.txt", "r")
    for line in file:
        print(line, end="") 
    
    file.close()
    print("\nFile opened and read successfully!")

except FileNotFoundError:

    print("Error: The file 'hello.txt' was not found. Please ensure it is in the correct directory.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
## create a function that that reads a csv with pandas .if filke is mission give keyerror 
def read_csv(filename):
    try:
        import pandas as pd 
        df = pd.read_csv(filename)
        return df 
    except KeyError:
        return "filenot found"
    except FileNotFoundError:
        return "filenot found"
filename = "data.csv"
print(read_csv(filename))
```

```python
# use assert to check that a list is not empty before computing its average !
def average(list):
    assert len(list) != 0 , "list is   empty"
    return sum(list) / len(list)
list = [1,2,3,4,5,6]
print(average(list))
```

```python
class DataValidationError(Exception):
    def __init__(self, message="Invalid data"):
        super().__init__(message)

def validate_data(data):
    
    if not isinstance(data, dict):
        raise DataValidationError("Data must be a dictionary.")

    if "name" not in data or not isinstance(data["name"], str):
        raise DataValidationError("Name must be a string.")

    if "age" not in data or not isinstance(data["age"], int) or data["age"] <= 0:
        raise DataValidationError("Age must be a positive integer.")

def process_data(data):

    print(f"Processing data: Name = {data['name']}, Age = {data['age']}")

# Example Usage
if __name__ == "__main__":
    valid_data = {"name": "Alex", "age": 30}
    invalid_data1 = ["Lucifero", 25]
    invalid_data2 = {"name": 123, "age": "thirty"}
    invalid_data3 = {"name": "Lucifero", "age": -5}

    try:
        validate_data(valid_data)
        process_data(valid_data)  # Only process if validation passes
    except DataValidationError as e:
        print(f"Validation Error: {e}")

    try:
        validate_data(invalid_data1)
        process_data(invalid_data1)
    except DataValidationError as e:
        print(f"Validation Error: {e}")

    try:
        validate_data(invalid_data2)
        process_data(invalid_data2)
    except DataValidationError as e:
        print(f"Validation Error: {e}")

    try:
        validate_data(invalid_data3)
        process_data(invalid_data3)
    except DataValidationError as e:
        print(f"Validation Error: {e}")
```
