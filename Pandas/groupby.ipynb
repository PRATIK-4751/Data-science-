```python
##group a df by a single coloumn and calculate mean of another column 
import pandas as pd 
df = pd.read_csv('iphones.csv')
df.groupby('product_name')['price_inr'].mean()
```

```python
##group by multiple coloumns and count the rowa in each group !
df.groupby(['product_name', 'source']).size()
```

```python
#find the maximum coloumns and count the rows in each group !

max_values = df.max()
group = df.groupby('product_name')
row_count = group.size()
print(max_values)
print(row_count)
```

```python
# find the maximum value in each group



max_values = group['price_inr'].max()
print(max_values)
```

```python
## calculate multiple aggregetions at once (mean,median,std) for each group !
agg = group['price_inr'].agg(['mean','median','std'])
print(agg)
```

```python
## use .size() and .count() and explain the differece! 
result = group['price_inr'].size()
print(result)
result = group['price_inr'].count()
print(result)
```

```python
## Apply a custom aggretion function to groups 
def custom_egg(x):
    return x.max() - x.min()

result = group['price_inr'].agg(custom_egg)
print(result)
```

```python
## get the row with largest value in each group
idx = df.groupby('product_name')['price_inr'].idxmax()
result = df.loc[idx]
print(result)
```

```python
## convert a grouped result back into DataFrame with .reset_index()
print(df)

```

```python
result = df.groupby("product_name")["price_inr"].sum()
result = result.reset_index()
result
```

```python
## perform group wise transformation (eg. subtract group mean from each value)
df['price_diff_from_mean'] = df['price_inr'] - df.groupby('source')['price_inr'].transform('mean')

# Display the result
df[['product_name', 'source', 'price_inr', 'price_diff_from_mean']].head(10)
```

```python
## group by and sort groups by aggregate value
df.groupby('source')['price_inr'].mean().sort_values(ascending=False)
```
