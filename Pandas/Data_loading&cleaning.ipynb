```python
## load a csv file and show 5 dataframe !
import pandas as pd 
df =pd.read_csv("iphones.csv")
df.head(5)
```

```python
# check  for missing values in the dataframe !
df.isnull().sum()
```

```python
## DROP ROWS WITH MISSING VALUES
df.dropna(inplace=True)
df.isnull().sum()
```

```python
## fill missing values and reset the index ! 
df.fillna("unknown",inplace=True)
df.reset_index(drop=True,inplace=True)
df
```

```python
## fill missng values in a coloumn with coloumn mean 

df['price_inr'].fillna(df['price_inr'].mean(), inplace=True)
df
```

```python
## renanme coloumns in a dataframe 
df.rename(columns={'price_inr':'price_inr_mean'}, inplace=True)
df
```

```python
## fill missng values in a coloumn with coloumn mean 

df['price_inr_mean'].fillna(df['price_inr_mean'].mean(), inplace=True)
df
```

```python
## load only specific columns from a dataframe
df = pd.read_csv('iphones.csv', usecols=['product_name'])
df
```

```python
# handle dupicate values in a dataframe
df.drop_duplicates(inplace=True)
df
```

```python
#Convert a column to the correct data type (e.g., string  datetime).
df['date'] = pd.to_datetime(df['date'])
df
```
