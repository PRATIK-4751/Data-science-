```python
import pandas as pd
df = pd.read_csv('iphones.csv')
df
```

```python
#Select a single column as a Series and multiple columns as a DataFrame.
df['price_inr']
df[['product_name','price_inr']]
```

```python
## select row by index lables (.loc) and iloc 
df.loc[0]
df.iloc[0]
```

```python
## filter rows where a coloumn's value meets a contion 
df[df['price_inr']>50]
```

```python
# select a range of rows using slcing  !
df[0:5]
```

```python
## check and reset the index  
df.index
df.reset_index(inplace=True)
df
```

```python
#set a coloumn as the index and sort dataframe by index ! 
df.set_index('product_name',inplace=True)
df.sort_index(inplace=True)
df
```

```python
#use .loc with multiple conditions And and Or 
df.loc[(df['price_inr']>50) & (df['price_inr']<100)]
df.loc[(df['price_inr']>50) | (df['price_inr']<100)]
```
